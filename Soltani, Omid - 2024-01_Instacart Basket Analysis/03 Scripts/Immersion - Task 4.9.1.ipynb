## 3. Import your the new customer data set as a dataframe.

## 4. Wrangle the data so that it follows consistent logic.

```python
# rename the columns "Surnam", "n_dependants" & "fam_status"

df_customers.rename(columns={'Surnam': 'Surname', 'n_dependants': 'number_of_dependents', 'fam_status': 'marital_status'}, inplace=True)

# check df dimensions to verify changes

df_customers.head()
```

```python
# check df_customers info

df_customers.info()
```

```python
# convert dtype for the column "date_joined" for easier date-related operations

df_customers['date_joined'] = pd.to_datetime(df_customers['date_joined'])

print(df_customers['date_joined'].dtype)
```

### Since there are 11,259 null values in the "First Name" column (as we can understand from the 'Non-Null Count' column above), and its content is likely unique for each user, it may not provide meaningful insights into customer behaviors laid out in the Project Brief, such as brand loyalty, ordering habits, region, age, marital status, income, and product preferences.

```python
# drop the column "First Name" 

df_customers.drop(columns=['First Name'], inplace=True)
```

```python
# check df_customers info

df_customers.info()
```

## 6. Combine your customer data with the rest of your prepared Instacart data.

```python
# import df_ords_prods

df_ords_prods = pd.read_pickle(os.path.join(path, '02 Data', '00 Prepared Data', 'ords_prods_end.pkl'))
```

```python
# check df_ord_prods info

df_ords_prods.info()
```

```python
# combine df_ords_prods (primary) and df_customers

df_combined = pd.merge(df_ords_prods, df_customers, on = 'user_id', how = 'left')
```

```python
# check df_combined info

df_combined.info()
```

```python
# check df_combined dimensions

df_combined.head(10)
```

```python
# replace positive and negative infinity values with NaN

df_combined.replace([np.inf, -np.inf], np.nan, inplace = True)
```
